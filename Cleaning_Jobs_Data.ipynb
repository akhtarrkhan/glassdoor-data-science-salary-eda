# Glassdoor Jobs Data Cleaning

## Objective
Clean and transform raw Glassdoor job posting data to prepare it for analysis.

## Dataset
- Source: Glassdoor job postings dataset
- Contains job titles, salary estimates, company info, and descriptions

## Tools Used
- Python
- Pandas

```python
## 1. Load Dataset
import pandas as pd
df = pd.read_csv('glassdoor_jobs.csv')
```

## 2. Initial Data Inspection
- Shape of dataset
- Column overview
- Missing values

```python
df.head()

#Cleaning tasks
#salary parsing
#company name
#state field
#age of company
#parsing of job description (python, etc)
```

```python
df = df[df['Salary Estimate'] != '-1']
pd.set_option('display.max_rows', None)
df.shape
pd.set_option('display.max_rows', None)
```

## 3. Salary Cleaning
- Remove invalid salary rows
- Extract min, max, and average salary
- Identify hourly vs yearly salaries
- Employer provided salary flag

```python
#Standardizing Salary column by removing texts and dollar signs/ salary parsing
salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
df.shape

salary
minus_kd = salary.apply(lambda x: x.replace('K', '').replace('$', ''))
minus_kd.head()

```

```python
#We still see that there are rows with hourly pay & employer provided salary hence creating columns for the same
df['hourly'] = df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
df['employer_provided'] = df['Salary Estimate'].apply(lambda x: 1 if 'employer provided salary' in x.lower() else 0)
df[df['hourly'] == 1].head(10)
```

```python
#creating a salary column with no texts
min_hr = minus_kd.apply(lambda x: x.lower().replace('per hour', '').replace('employer provided salary:', ''))
min_hr.head()
```

```python
#Creating minimum and maximum salaries columns in the main df basis the prev column
df['min_salary'] = min_hr.apply(lambda x: int(x.split('-')[0]))
df['min_salary'].dtype
df['max_salary'] = min_hr.apply(lambda x: int(x.split('-')[1]))
df.head()
```

```python
df['avg_salary'] = (df.max_salary + df.min_salary)/2
df.head()
```

## 4. Company Name Cleaning
- Remove rating text from company names

```python
#Standardizung company name to include texts only
#Numerical ratings are present at the end of company names(where rating > 0) which are 3 characters long.
#Removing the last 3 chars seems to be the easiest approach

df['company_txt'] = df.apply(lambda x: x['Company Name'] if x['Rating'] < 0 else x['Company Name'][:-3], axis = 1)
df.head()
```

## 5. Location Feature Engineering
- Extract state from location

```python
#State Field
df['job_state'] = df['Location'].apply(lambda x: x.split(',')[1])
df.head()
df.job_state.value_counts()
```

```python
#Creating an indicator for jobs in the same location as company headquarters
df.columns
df['same_state'] = df.apply(lambda x: 1 if x['Location'] == x['Headquarters'] else 0, axis = 1)
df.head()
```

## 6. Company Age Feature
- Calculate company age from founding year

```python
#age of company
df['age'] = df['Founded'].apply(lambda x: x if x < 0 else 2026 - x)
df.head()
```

## 7. Job Description Feature Engineering
- Calculate description length
- Extract skill keywords (Python, R, SQL, etc.)

```python
#Parsing of job description (Python etc)
df['Job Description']

df['python_yn'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
df['python_yn'].value_counts()
df['R_yn'] = df['Job Description'].apply(lambda x: 1 if 'r studio' in x.lower() or 'r-studio' in x.lower() else 0)
df['sql'] = df['Job Description'].apply(lambda x: 1 if 'sql' in x.lower() else 0)
df['spark'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0)
df['aws'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
df['excel'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
```

```python
#dropping the first unnamed columm
df_out = df.drop(['Unnamed: 0'], axis = 1)
```

## 8. Final Dataset Export
- Save cleaned dataset for analysis

```python
#Exporting the final cleaned data
df_out.to_csv('salary_data_cleaned.csv', index = False)
```
